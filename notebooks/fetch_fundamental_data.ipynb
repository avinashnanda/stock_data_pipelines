```python
from screener_client.fetch import fetch_all_data

BASE_URL = "https://www.screener.in/company/DIXON/consolidated/"

data = await fetch_all_data(BASE_URL)
```

```python
import json
import pandas as pd
import numpy as np

def converter(o):
    if isinstance(o, (pd.Timestamp, pd.Timedelta)):
        return o.isoformat()
    if isinstance(o, (np.int64, np.int32, int)):
        return int(o)
    if isinstance(o, (np.float64, np.float32, float)):
        return float(o)
    return str(o)

with open("output.json", "w") as f:
    json.dump(data, f, indent=4, default=converter)
```
