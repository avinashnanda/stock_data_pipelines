### First connect:

```python
import duckdb
con = duckdb.connect("db/screener_financials.duckdb", read_only=True)
```

### Show tables:

```python
con.sql("""
SHOW TABLES;
""").df()
```

### Count rows:

```python
con.sql("""
SELECT 'companies' AS table_name, COUNT(*) AS row_count FROM companies
UNION ALL
SELECT 'raw_company_json', COUNT(*) FROM raw_company_json
UNION ALL
SELECT 'failed_companies', COUNT(*) FROM failed_companies;
""").df()
```

### List companies:

```python
con.sql("""
SELECT *
FROM companies
ORDER BY company_name
LIMIT 100;
""").df()
```

### Latest snapshots:

```python
con.sql("""
SELECT company_id, source_url, scraped_at
FROM raw_company_json
ORDER BY scraped_at DESC
LIMIT 50;
""").df()
```

### Latest failures:

```python
con.sql("""
SELECT *
FROM failed_companies
ORDER BY last_attempt DESC
LIMIT 50;
""").df()
```

### Find one symbol:

```python
con.sql("""
SELECT company_id, source_url, scraped_at
FROM raw_company_json
WHERE UPPER(source_url) LIKE '%/COMPANY/AARON/%'
ORDER BY scraped_at DESC;
""").df()
```

### Top-level JSON keys:

```python
con.sql("""
SELECT
  company_id,
  scraped_at,
  json_keys(payload_json) AS top_level_keys
FROM raw_company_json
ORDER BY scraped_at DESC
LIMIT 2;

""").df()
```

### One full payload preview:



```python
con.sql("""
SELECT payload_json
FROM raw_company_json
WHERE UPPER(source_url) LIKE '%/COMPANY/AARON/%'
ORDER BY scraped_at DESC
LIMIT 1;
""")
```

### Stored chart keys:

```python
con.sql("""
SELECT
  json_keys(json_extract(payload_json, '$.charts')) AS chart_keys
FROM raw_company_json
WHERE UPPER(source_url) LIKE '%/COMPANY/AARON/%'
ORDER BY scraped_at DESC
LIMIT 1;
""")
```

### Stored schedule keys:

```python
con.sql("""
SELECT
  json_keys(json_extract(payload_json, '$.schedules')) AS schedule_keys
FROM raw_company_json
WHERE UPPER(source_url) LIKE '%/COMPANY/AARON/%'
ORDER BY scraped_at DESC
LIMIT 1;

""")
```

### Extract summary fields:



```python
con.sql("""
SELECT
  json_extract_string(payload_json, '$.summary.company_name') AS company_name,
  json_extract_string(payload_json, '$.summary.Market Cap') AS market_cap,
  json_extract_string(payload_json, '$.summary.Current Price') AS current_price,
  json_extract_string(payload_json, '$.summary.Stock P/E') AS stock_pe
FROM raw_company_json
WHERE UPPER(source_url) LIKE '%/COMPANY/AARON/%'
ORDER BY scraped_at DESC
LIMIT 1;
""")
```

### Extract chart JSON:

```python
con.sql("""
SELECT
  json_extract(payload_json, '$.charts.price_dma_volume') AS price_dma_volume
FROM raw_company_json
WHERE UPPER(source_url) LIKE '%/COMPANY/AARON/%'
ORDER BY scraped_at DESC
LIMIT 1;
""")
```

### Extract quarterly sales schedule JSON:

```python
con.sql("""
SELECT
  json_extract(payload_json, '$.schedules.sales_quarterly') AS sales_quarterly
FROM raw_company_json
WHERE UPPER(source_url) LIKE '%/COMPANY/RELIANCE/%'
ORDER BY scraped_at DESC
LIMIT 1;
""")
```

### Find symbols that have stored snapshots:

```python
con.sql("""
SELECT DISTINCT
  regexp_extract(UPPER(source_url), '/COMPANY/([^/]+)/', 1) AS symbol
FROM raw_company_json
WHERE source_url IS NOT NULL
ORDER BY symbol;
""")
```

### Latest snapshot per symbol:



```python
con.sql("""
WITH ranked AS (
  SELECT
    regexp_extract(UPPER(source_url), '/COMPANY/([^/]+)/', 1) AS symbol,
    company_id,
    source_url,
    scraped_at,
    ROW_NUMBER() OVER (
      PARTITION BY regexp_extract(UPPER(source_url), '/COMPANY/([^/]+)/', 1)
      ORDER BY scraped_at DESC
    ) AS rn
  FROM raw_company_json
)
SELECT symbol, company_id, source_url, scraped_at
FROM ranked
WHERE rn = 1
ORDER BY symbol;
""")
```
